```python
import cycler as cycler
import numpy as np

from matplotlib import pyplot as plt
from IPython import display
```

```python
import matplotlib as mpl

from pandas import plotting

_FIGURE_DPI = 288
_PLOT_DEFAULT_CMAP = 'Set2'
_PLOT_TEXT_COLOR = '#444444'
_PLOT_TICK_COLOR = '#666666'

def set_default_styles(retina: bool = True) -> None:
  plotting.register_matplotlib_converters()

  def set_retina():
    display.set_matplotlib_formats('retina')
    return _FIGURE_DPI // 2

  # Import and register fonts with Matplotlib so we can use them.
  # import_default_fonts()
  dpi = set_retina() if retina else _FIGURE_DPI

  mpl_styles = {
      'axes.axisbelow': True,
      'axes.edgecolor': _PLOT_TICK_COLOR,
      'axes.grid': True,
      'axes.labelcolor': _PLOT_TEXT_COLOR,
      # Set default color palette.
      'axes.prop_cycle': cycler.cycler('color', mpl.cm.get_cmap(_PLOT_DEFAULT_CMAP).colors),
      'axes.spines.right': True,
      'axes.spines.top': True,
      # 'font.family': ['Poppins', 'Noto Sans', 'Roboto'],
      # You can scale all other text by scaling this.
      'font.size': 9.0,
      # 'font.weight': 'light',
      'figure.figsize': (6, 4),
      # Up the default resolution for figures.
      'figure.dpi': dpi,
      'savefig.dpi': dpi,
      'lines.linewidth': 1.25,
      'grid.color': '#EEEEEE',
      'grid.linewidth': 0.4,
      'text.color': _PLOT_TEXT_COLOR,
      'xtick.color': _PLOT_TICK_COLOR,
      'ytick.color': _PLOT_TICK_COLOR,
      # 'xtick.alphas': 0.85,
      'legend.borderaxespad': 1.0,
      'legend.borderpad': 0.8,
      'legend.edgecolor': '1.0',
      'legend.facecolor': '#F3F3F3',
      'legend.fontsize': 'small',
      # 'legend.framealpha': 0.75,
      'legend.framealpha': 1.0,

      'legend.labelspacing': 0.6,
  }

  mpl.rcParams.update(mpl_styles)
  
set_default_styles()
```

```python
r = np.arange(0, 4, 0.2)
theta = 2 * np.pi * r

l0_r = np.array(list(map(lambda item: item[1], filter(lambda item: item[0] % 2 == 0, enumerate(r)))))
l0_theta = 2 * np.pi * l0_r

l1_r = np.array(list(map(lambda item: item[1], filter(lambda item: item[0] % 5 == 0, enumerate(r)))))
l1_theta = 2 * np.pi * l1_r

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.plot(l0_theta, l0_r, marker='o')
ax.plot(l1_theta, l1_r, marker='s')
ax.set_rmax(4)
ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```

```python
l0_nodes = np.arange(0, r.shape[0], 1)
l1_nodes = l0_nodes[l0_nodes % 2 == 0]
l2_nodes = l0_nodes[l0_nodes % 5 == 0]

plt.plot(l0_nodes, [4]*l0_nodes.shape[0], '-s')
plt.plot(l1_nodes, [5]*l1_nodes.shape[0], '-s')
plt.plot(l2_nodes, [6]*l2_nodes.shape[0], '-s')
plt.gca().set_yticks(range(10))
```
