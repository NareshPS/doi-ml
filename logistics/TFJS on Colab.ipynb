# Example of usage

**Remember not to use `const` or `let`! Use `var` instead**

This is how you can execute shell commands:

```
var { spawn } = require('child_process');
var sh = (cmd) => { 
    $$.async();
    var sp = spawn(cmd, { cwd: process.cwd(), stdio: 'pipe', shell: true, encoding: 'utf-8' });
    sp.stdout.on('data', data => console.log(data.toString()));
    sp.stderr.on('data', data => console.error(data.toString()));
    sp.on('close', () => $$.done());
};
var run_async = async (pf) => {
  $$.async();
  await pf();
  $$.done();
};
sh('npm init -y');
```

```
var tf = require('@tensorflow/tfjs-node-gpu');
console.log(tf.backend().isUsingGpuDevice);
```

```
var tensorToSave = tf.tensor([[1, 2], [3, 4]]);
tensorToSave.print()
```

```
// saving tensor
var fs = require('fs');
fs.writeFileSync('saved-tensor.json', JSON.stringify(tensorToSave.arraySync()), 'utf-8');
sh('ls -la saved-tensor.json');
```

```
// loading tensor
var fs = require('fs');
var str = fs.readFileSync('saved-tensor.json', 'utf-8');
var tensorRestored = tf.tensor(JSON.parse(str));
tensorRestored.print()
```
