```python
import nlpaug.augmenter.word as naw
```

# Functions

```python
def backtranslation(text):
  aug = naw.BackTranslationAug(
      from_model_name='facebook/wmt19-en-de', 
      to_model_name='facebook/wmt19-de-en'
  )
  
  return aug.augment(text)
```

# Back-Translation

```python
timeit backtranslation('The whale is jumping in the water')
```

```python
timeit backtranslation(['The whale is jumping in the water', 'foo bar', 'xyz abc', 'abc', 'dskfdsf', 'asdkfjdk', 'dsfdsf', 'dfdsf dfsdsf', 'asdasd', 'sadsad'])
```
