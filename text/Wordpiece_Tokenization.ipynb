```python
# ! pip install tensorflow-datasets
import tensorflow_datasets as tfds

train_ds, = tfds.load(
    'wikipedia_toxicity_subtypes',
    split=['train[:30%]'],
    as_supervised=True
)
```

```python
from nltk.tokenize import WordPunctTokenizer
     
# Create a reference variable for Class WordPunctTokenizer
tk = WordPunctTokenizer()
```

```python
from itertools import combinations
from collections import Counter

def make_subsequences(s):
    return [s[x:y] for x,y in combinations(range(len(s) + 1), r = 2)]

def short_subsequences(s, limit=10):
    subs = make_subsequences(s)
    short_subs = filter(lambda s: len(s) <= limit, subs[len(s):])
    return subs[:len(s)] + list(short_subs)

def count_subs(tokens, counter=Counter()):
    for token in tokens:
        token = token.lower()
        subs = short_subsequences(token)
        counter.update(subs)
    
    return counter
```

```python
## Adaptation

counter = Counter()
for item in train_ds.as_numpy_iterator():
    tokens = tk.tokenize(item[0].decode('utf-8'))
    count_subs(tokens, counter)

counter.most_common(100)
```
